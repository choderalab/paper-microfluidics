```python
import plate
from random import randint
from IPython.display import SVG, display_svg
```

```python
plate96 = plate.Plate96(barrier_tickness=0.5, well_diameter=7.0, margin=0.5)
display_svg(SVG(plate96.svg.tostring()))
```

```python
plate384 = plate.Plate384(barrier_tickness=0.5, well_diameter=3.5, margin=0.5)
display_svg(SVG(plate384.svg.tostring()))
```

```python
colors = [['rgb({},{},{})'.format(randint(0,255), randint(0,255), randint(0,255))
           for j in range(24)] for i in range(16)]

cheerful_plate = plate.Plate384(barrier_tickness=0.5, well_diameter=3.5, margin=0.5, well_fill=colors)
display_svg(SVG(cheerful_plate.svg.tostring()))
```
